# Problema de la Mochila 0/1

En esta sección se muestran ejemplos de uso de:

- Greedy
- Backtracking
- Recocido Simulado

```python
# Generar instancia aleatoria

valores, pesos, capacidad = generar_instancia_mochila(n_objetos=10)

print("Valores:", valores)
print("Pesos:", pesos)
print("Capacidad:", capacidad)
```

## Solución con Greedy

El algoritmo greedy selecciona objetos utilizando la mejor relación valor/peso.

```python
seleccion, valor = mochila_greedy(valores, pesos, capacidad)

print("Selección:", seleccion)
print("Valor total:", valor)
```

## Solución exacta con Backtracking

El algoritmo explora combinaciones posibles utilizando poda Branch and Bound.

```python
valor_optimo, peso_optimo, seleccion_optima = mochila_backtracking(
    pesos,
    valores,
    capacidad
)

print("Valor óptimo:", valor_optimo)
print("Peso óptimo:", peso_optimo)
print("Selección:", seleccion_optima)
```

## Recocido Simulado

Metaheurística inspirada en procesos físicos de enfriamiento.

```python
solucion_rs, valor_rs = mochila_recocido_simulado(
    valores,
    pesos,
    capacidad
)

print("Solución:", solucion_rs)
print("Valor:", valor_rs)
```

# Problema del Agente Viajero (TSP)

En esta sección se muestran ejemplos utilizando generación de rutas aleatorias y optimización con 2-OPT.

```python
# Generar matriz aleatoria
matriz = generar_matriz_aleatoria(8)

print("Matriz de costos:")

for fila in matriz:
    print(fila)

# Crear instancia
tsp = TSP(matriz)

# Generar ruta aleatoria
ruta = tsp.generar_ruta()

print("\nRuta inicial:", ruta)

# Calcular costo inicial
costo = tsp.calcular_costo(ruta)

print("Costo inicial:", costo)
```

## Optimización con 2-OPT

El algoritmo 2-OPT mejora iterativamente una ruta intercambiando segmentos.

```python
nueva_ruta, nuevo_costo = tsp.dos_opt(ruta)

print("Ruta optimizada:", nueva_ruta)
print("Nuevo costo:", nuevo_costo)
```
